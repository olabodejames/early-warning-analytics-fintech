
## RQ3/RQ4 panel build



```python
from src.config import EPAYMENTS_CSV, FRAUD_SUMMARY_CSV, MERGED_PANEL_CSV, FEATURE_MATRIX_CSV
from src.io_utils import read_csv, write_csv
from src.cleaning import clean_epayments_long, clean_fraud_summary
from src.feature_engineering import build_annual_totals, merge_payments_fraud, add_growth_rates, build_rq4_feature_matrix

epay = clean_epayments_long(read_csv(EPAYMENTS_CSV))
fraud = clean_fraud_summary(read_csv(FRAUD_SUMMARY_CSV))

annual_pay = build_annual_totals(epay)
panel = merge_payments_fraud(annual_pay, fraud)
panel = add_growth_rates(panel)

write_csv(panel, MERGED_PANEL_CSV, index=False)

X = build_rq4_feature_matrix(panel)
write_csv(X, FEATURE_MATRIX_CSV, index=False)

panel.head(), X.head()
```
